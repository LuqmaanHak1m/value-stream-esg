```python
import sys
import subprocess
from bs4 import BeautifulSoup

def get_html(url):
    result = subprocess.run(
        [sys.executable, "webscrape.py", "--url", url],
        capture_output=True,
        text=True
    )

    if result.returncode != 0:
        raise RuntimeError("webscrape.py failed")

    with open("page.html", encoding="utf-8") as f:
        soup = BeautifulSoup(f.read(), "html.parser")

    return soup

```

# Function

```python
def get_title_and_dates(articles):
    dates = []
    titles = []

    for article in articles:
        titles.append(article.a.text.strip())
        dates.append(article.time.text.strip())

    clean_dates = [datetime.datetime.strptime(date, "%B %d, %Y") for date in dates]

    return titles, clean_dates
```

```python
def get_introductions(page_links):
    first_line_list = []
    for page in page_links:
        page_soup = get_html(page)

        article = page_soup.find("article")

        for p in article.find_all("p"):
            text = p.get_text(" ", strip=True)
            if text:
                first_non_empty = p
                break
        
        if first_non_empty:
            first_line_list.append(first_non_empty.text)

    return first_line_list
```

```python
def scrape_news(search_keyword, past_x_years):
    min_year = datetime.datetime.today().year - past_x_years

    base_url = "https://esgnews.com/?s="
    sorting = "&orderby=newest&order=DESC"
    query = search_keyword.replace(" ", "+")

    full_url = base_url + query + sorting
    soup = get_html(full_url)

    articles = soup.find_all("article")

    # titles, dates = get_title_and_dates(articles)

    # cut_index = len(dates)

    # for i, d in enumerate(dates):
    #     if d.year <= min_year:
    #         cut_index = i
    #         break

    # titles = titles[:cut_index]
    # dates = dates[:cut_index]

    # print(titles)
    # print(dates)

    page_links = []
    for article in articles:
        if article.a:
            page_links.append(article.a.get("href"))

    stuff = get_introductions(page_links)

    # return zip(titles, dates)
    return stuff
```
