### 2. More advanced notions 🚀

Before we go further, here's a handy cheat-sheet for some of the most used BeautifulSoup methods & properties:

|Method / Property|What it does|Result Type|Example|
|---|---|---|---|
|`.find()`|Retrieves the first matching tag|Tag object|`soup.find('div', class_='header')`|
|`.find_all()`|Retrieves all matching tags as a list|ResultSet (list)|`soup.find_all('a')`|
|`.select_one()`|Finds the first match using CSS Selectors|Tag object|`soup.select_one('#main-title')`|
|`.select()`|Finds all matches using CSS Selectors|List of Tags|`soup.select('ul > li.item')`|
|`.find_parent()`|Moves up the tree to find the nearest parent|Tag object|`tag.find_parent('div')`|
|`.find_next_sibling()`|Grabs the next element at the same level|Tag object|`tag.find_next_sibling('p')`|
|`.get_text()`|Strips HTML and returns only the inner text|String|`soup.find('h1').get_text()`|
|`.get()`|Retrieves the value of a specific attribute|String|`link.get('href')`|
